```python
import pandas as pd 
import geopandas as gpd
import folium
import folium
from folium.plugins import HeatMap
import numpy as np
import branca
import branca.colormap as cm
```

```python
colombia_depto = gpd.read_file('depto.zip')
```

```python
santander = colombia_depto[colombia_depto.DPTO == '68' ]
#colombia_depto
```

```python
santander_map = folium.Map(location=[7.12,-73], zoom_start=9, tiles='cartodbpositron' )
```

```python
print('default map crs: ',santander_map.crs)
```

```python
santander = santander.to_crs({'init':'epsg:3857'})  
```

```python
folium.GeoJson(santander).add_to(santander_map)
santander_map
```

```python
for_map = pd.read_csv('campaign.tsv', sep='\t')

for_map.sort_values('Amount', ascending=False).head()
```

```python
hmap = folium.Map(location=[7.12,-73.11], zoom_start=14.5, )

hm_wide = HeatMap( list(zip(for_map.lat.values, for_map.lon.values, for_map.Amount.values)), 
#                  gradient={0.4: 'blue', 0.65: 'yellow', 1: 'red'},
#                   min_opacity=0.2,
#                   radius=15, blur=15, 
#                   max_zoom=1, 
                 )
colormap = cm.LinearColormap(colors=['darkblue', 'blue', 'cyan', 'yellow', 'orange', 'red'],
                             index=[0, 25, 50, 100, 200, 400], vmin=0, vmax=400,
                             caption='Absolute value')
#folium.GeoJson(santander).add_to(hmap)
hmap.add_child(hm_wide)
hmap.add_child(colormap)
```
