```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.signal import savgol_filter
import datetime
from mpl_toolkits.basemap import Basemap
import xarray as xr
%matplotlib inline
```

```python
plt.figure(figsize=(10,8))
map = Basemap(llcrnrlon=-74.6, llcrnrlat=5.6, urcrnrlon=-72., urcrnrlat=8.2,resolution="i")
parallels=np.arange(5.5,8.5,0.5)
map.drawparallels(parallels,labels=[True,False,False,True])
meridians = np.arange(-74.5,-71.5,0.5)
map.drawmeridians(meridians,labels=[False,True,True,False])
map.drawmapboundary(fill_color='#FAD7A0')
map.drawcountries()
map.drawstates()
map.fillcontinents(color='#cc3945', lake_color='aqua', zorder = 0)
map.drawcoastlines(color = '0.15')

plt.scatter(-72.73,6.7,c='k',label="Malaga's station")
plt.text(-72.73+0.05,6.7,s="Malaga",fontsize=14)
plt.scatter(-73.13,7.12,c='b',label="Bucaramanga's City")
plt.text(-73.13-0.65,7.12,s="Bucaramanga",fontsize=14,color='b')

plt.text(-74.25,6.25,s="Department of Santander",fontsize=16,color='r',weight='bold')
plt.legend(loc=1)
plt.show()
```

```python
plt.figure(figsize=(10,8))
map = Basemap(llcrnrlon=-72.79, llcrnrlat=6.64, urcrnrlon=-72.68, urcrnrlat=6.75,resolution="i")
parallels=np.arange(6.65,6.75,0.01)
map.drawparallels(parallels,labels=[True,False,False,True],rotation=0)
meridians = np.arange(-72.78,-72.68,0.01)
map.drawmeridians(meridians,labels=[False,True,True,False],rotation=90)
map.drawmapboundary(fill_color='#FAD7A0')
map.drawcountries()
map.drawstates()
map.fillcontinents(color='#cc3945', lake_color='aqua', zorder = 0)
map.drawcoastlines(color = '0.15')
y = np.arange(6.65,6.75,0.01)
x = np.arange(-72.78,-72.68,0.01)
x,y=np.meshgrid(x,y)
x,y=map(x,y)

map.scatter(x,y,c='b',label='grid points IRI')
plt.scatter(-72.73,6.7,c='k',label="Malaga's station")


plt.legend(loc=1)
plt.show()
```
